```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("/content/drive/MyDrive/factcheckAI/data/processed/truthfulqa_working.csv")

df.head()
```

```python
#testing on the first 10 questions
test_df = df.head(10)

test_df
```

```python
#configuring Gemini
import google.generativeai as genai

```

```python
from google.colab import userdata

GEMINI_API_KEY = userdata.get("Gemini_API_key")
GROQ_API_KEY = userdata.get("Groq_API_key")
DEEPSEEK_API_KEY = userdata.get("deepseek_API_key")
```

```python


genai.configure(api_key=GEMINI_API_KEY)
```

```python
#creating the model for Gemini
gemini_model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
#testing gemini
response = gemini_model.generate_content("What is the capital of France?")

print(response.text)
```

```python
#configuring Groq
from groq import Groq

client = Groq(api_key=GROQ_API_KEY)
```

```python
#testing Groq
response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        { "role":"user","content":"What is the capital of France?" }
    ]
)

print(response.choices[0].message.content)
```

```python
#configuring deepseek
from openai import OpenAI
```

```python
deepseek = OpenAI(api_key=DEEPSEEK_API_KEY,base_url="https://api.deepseek.com")
```

```python
response = deepseek.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {
            "role":"user",
            "content":"What is the capital of France?"
        }
    ]
)

print(response.choices[0].message.content)
```

```python
#creating wrapper functions
#Gemini

def ask_gemini(question):
    response = gemini_model.generate_content(question)
    return response.text
```

```python
#Groq
def ask_groq(question):
    response = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role":"user",
                "content":question
            }])

    return response.choices[0].message.content
```

```python
#deepseek
def ask_deepseek(question):
    response = deepseek.chat.completions.create(
        model="deepseek-chat",
        messages=[
            {
                "role":"user",
                "content":question }
]
    )

    return response.choices[0].message.content
```

```python
#testing wrapper functions
print( ask_gemini("Who discovered gravity?"))
```

```python
print( ask_groq("Who discovered gravity?"))
```

```python
print( ask_deepseek("Who discovered gravity?"))
```

```python
#generating responses for 10 questions
responses = []

for _, row in tqdm(test_df.iterrows(), total=len(test_df)):
    question = row["Question"]

    responses.append({
        "Question": question,
        "Ground_Truth": row["Ground_Truth"],
        "Category": row["Category"],
        "Model": "Gemini",
        "Response": ask_gemini(question)})

    responses.append({
        "Question": question,
        "Ground_Truth": row["Ground_Truth"],
        "Category": row["Category"],
        "Model": "Groq-Llama3.3",
        "Response": ask_groq(question)})

    responses.append({
        "Question": question,
        "Ground_Truth": row["Ground_Truth"],
        "Category": row["Category"],
        "Model": "DeepSeek",
        "Response": ask_deepseek(question)})
```

```python
responses_df = pd.DataFrame(responses)

responses_df.head()
```

```python
#saving the results
responses_df.to_csv("/content/drive/MyDrive/factcheckAI/outputs/responses_test.csv",index=False)
```
