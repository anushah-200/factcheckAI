```python
import pandas as pd
import numpy as np
from datasets import load_dataset
```

```python
dataset = load_dataset("truthfulqa/truthful_qa", "generation")
```

```python
df = dataset["validation"].to_pandas()
```

```python
#creating working dataset
working_df = df[['type', 'category', 'question', 'best_answer']].copy()
```

```python
#renaming and making columns easier to understand
working_df.rename(columns={
    'type': 'Type',
    'category': 'Category',
    'question': 'Question',
    'best_answer': 'Ground_Truth'
}, inplace=True)
```

```python
#saving the raw dataset(will not be altered)
df.to_csv("/content/drive/MyDrive/factcheckAI/data/raw/truthfulqa_raw.csv",index=False)
```

```python
#saving the working dataset
working_df.to_csv("/content/drive/MyDrive/factcheckAI/data/processed/truthfulqa_working.csv",index=False)
```
